```python
# --- Baseline: scikit-learn LinearRegression vs. XGBoost XGBRegressor ---
from sklearn.datasets import load_diabetes  # offline alternative to fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import time

from xgboost import XGBRegressor

# Load dataset (replace with fetch_california_housing() if internet/cache available)
data = load_diabetes()
X, y = data.data, data.target

# Scale features (useful for LinearRegression; not strictly required for trees)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ----- Model 1: scikit-learn Linear Regression -----
t0 = time.perf_counter()
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_train_time = time.perf_counter() - t0

t0 = time.perf_counter()
y_pred_lr = lr.predict(X_test)
lr_pred_time = time.perf_counter() - t0

mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

# ----- Model 2: XGBoost Regression (replaces sklearn regression model) -----
t0 = time.perf_counter()
xgb = XGBRegressor(
    objective="reg:squarederror",
    n_estimators=600,
    learning_rate=0.05,
    max_depth=3,
    subsample=1.0,
    colsample_bytree=1.0,
    random_state=42,
    n_jobs=4
)
xgb.fit(X_train, y_train)
xgb_train_time = time.perf_counter() - t0

t0 = time.perf_counter()
y_pred_xgb = xgb.predict(X_test)
xgb_pred_time = time.perf_counter() - t0

mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)

print("LinearRegression  | MSE:", mse_lr,  "| R2:", r2_lr,
      "| train(s):", lr_train_time, "| predict(s):", lr_pred_time)

print("XGBRegressor      | MSE:", mse_xgb, "| R2:", r2_xgb,
      "| train(s):", xgb_train_time, "| predict(s):", xgb_pred_time)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor

# Load dataset
data = load_diabetes()
X, y = data.data, data.target

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Train XGBoost model
model = XGBRegressor(
    objective="reg:squarederror",
    n_estimators=600,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("R²:", r2)

```
