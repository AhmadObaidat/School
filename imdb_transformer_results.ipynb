```python
!pip install transformers datasets scikit-learn pandas

import pandas as pd
from datasets import load_dataset
from transformers import pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load dataset
dataset = load_dataset("imdb")
df_test = pd.DataFrame(dataset["test"])

# Sample data
sample_size = 100
df_sample = df_test.sample(n=sample_size, random_state=42).reset_index(drop=True)

# Load model
classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

# Run predictions
predictions = []
for text in df_sample["text"]:
    result = classifier(text[:512])[0]
    predictions.append(result["label"])

df_sample["predicted_label_text"] = predictions

# Convert labels
label_map = {
    "NEGATIVE": 0,
    "POSITIVE": 1
}
df_sample["predicted_label"] = df_sample["predicted_label_text"].map(label_map)

# Evaluate
y_true = df_sample["label"]
y_pred = df_sample["predicted_label"]

accuracy = accuracy_score(y_true, y_pred)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=["negative", "positive"]))

print("\nConfusion Matrix:")
print(confusion_matrix(y_true, y_pred))

# Correct and incorrect examples
correct_examples = df_sample[df_sample["label"] == df_sample["predicted_label"]]
incorrect_examples = df_sample[df_sample["label"] != df_sample["predicted_label"]]

print("\nNumber of correct predictions:", len(correct_examples))
print("Number of incorrect predictions:", len(incorrect_examples))

print("\n--- Successful Examples ---")
for i in range(min(3, len(correct_examples))):
    row = correct_examples.iloc[i]
    print(f"\nExample {i+1}")
    print("True Label:", "positive" if row["label"] == 1 else "negative")
    print("Predicted Label:", row["predicted_label_text"])
    print("Review Snippet:", row["text"][:500])
    print("-" * 80)

print("\n--- Failure Examples ---")
for i in range(min(3, len(incorrect_examples))):
    row = incorrect_examples.iloc[i]
    print(f"\nError Example {i+1}")
    print("True Label:", "positive" if row["label"] == 1 else "negative")
    print("Predicted Label:", row["predicted_label_text"])
    print("Review Snippet:", row["text"][:500])
    print("-" * 80)

# Save results
df_sample.to_csv("imdb_transformer_results.csv", index=False)
print("\nResults saved to imdb_transformer_results.csv")
```
